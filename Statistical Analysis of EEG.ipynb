```python
# Selecting the data of a single channel

chann = data[:,4]
```

```python
plt.plot(chann)
```

```python
std = []
for d in range(2,int(chann.shape[0]/2)):
    scope = 0
    for i in range(0, chann.shape[0],d):
        scope += chann[i: i+d].var()
        scope /=d
        
        std.append(scope)
        
    
```

```python
plt.plot(std)
```
